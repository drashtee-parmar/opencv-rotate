```python
import cv2
from matplotlib import pyplot as plt
import imutils
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
image = cv2.imread("opencv_logo.png")
plt_imshow("CV2-Image", image)
```

## grab the dimensions of the image and calculate the center of the image

```python
(h,w) = image.shape[:2]
(cX, cY) = (w // 2, h // 2)
```

```python
print(cX, cY)
```

```python
# rotate our image by 45 degrees around the center of the image
# positive value for anti clockwise rotation
# negative value for clockwise rotation
M = cv2.getRotationMatrix2D((cX, cY), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by 45 Degrees", rotated)
```

```python
# rotate our image by -90 degrees around the image
M = cv2.getRotationMatrix2D((cX, cY), -90, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by -90 Degrees", rotated)
```

```python
# rotate our image around an arbitrary point rather than the center
M = cv2.getRotationMatrix2D((10, 10), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by Arbitrary Point", rotated)
```

# imutils.rotate : rotate an image with the given degree

```python
# use our imutils function to rotate an image 180 degrees
rotated = imutils.rotate(image, 180)
plt_imshow("Rotated by 180 Degrees", rotated)
```

```python
# use our imutils function to rotate an image 90 degrees
rotated = imutils.rotate(image, 90)
plt_imshow("Rotated by 90 Degrees", rotated)
```

```python
# use our imutils function to rotate an image 180 degrees
rotated = imutils.rotate(image, -33)
plt_imshow("Original", image)
plt_imshow("Rotated by -33 Degrees", rotated)
```

# imutils.rotate_bound : ensuring the entire rotated image still renders within the viewing area

```python
# rotate our image by 33 degrees counterclockwise, ensuring the
# entire rotated image still renders within the viewing area
rotated = imutils.rotate_bound(image, -33)
plt_imshow("Rotated Without Cropping", rotated)
```
